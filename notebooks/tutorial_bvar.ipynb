# Bayesian VAR tutorial

```python
from MacroPy import load_datasets

# Load the Stock & Watson (2001) dataset
df_sw = load_datasets(name="SW2001")
df_sw
```

```python
from MacroPy import generate_series_plot

# Generate a series plot for the Stock & Watson (2001) dataset
series_titles = ["Inflation", "Unemployment Rate", "Fed Funds Rate"]
generate_series_plot(df_sw, series_titles=series_titles)
```

```python
from MacroPy import BayesianVAR

bvar = BayesianVAR(df_sw, lags=4, hor=24, irf_1std=0)
bvar.model_summary()
```

```python
bvar.sample_posterior(plot_coefficients=True)
```

```python
irfs = bvar.compute_irfs(plot_irfs=True)
```
